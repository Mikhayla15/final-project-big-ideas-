```python
import pandas as pd
data = pd.read_csv("dataset.csv")
data.head(100)
```

this code drops all the columns that contain the offences

```python
data[data['OFFENSES'].notna()]
```

This code filters out the data so that only the engage in fihgting citations show 

```python
offense_data = data[data['OFFENSES'].str.contains('ENGAGE IN FIGHTING', case=False, na=False)]
print(offense_data)
```

```python
offense_data_filtered = offense_data[data['NEIGHBORHOOD'].notna()]
print(offense_data_filtered)
```

This code prints the number of engage in fighting in each neighborhood

```python
neighborhood_counts = offense_data['NEIGHBORHOOD'].value_counts()
print(neighborhood_counts)
```

This code groups the data by the "NEIGHBORHOOD" column and then count the number of occurances for each neighborhood. After that, it plots a bar chart using the counts. 

```python
import matplotlib.pyplot as plt

realData = offense_data.groupby("NEIGHBORHOOD").count()

realData["OFFENSES"].plot.bar(figsize=(20, 20), title="Engage In Fighting Offenses")
plt.xlabel("Neighborhood")
plt.ylabel("Count of Offenses")
plt.show()  # Show the plot
```

The code below arranges the fighting complaints in ascending order 

```python
sortedneighborhood=realData.sort_values("OFFENSES")
sortedneighborhood["OFFENSES"].plot.bar(figsize=(20,20), title="Engage In Fighting Offenses")
```

As we can see from the graphs above, South Side Flats has the most highest number of offenses that engage in fighting while the west end has the least amount of offenses.
